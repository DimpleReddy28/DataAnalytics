```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("class_marks.csv")
df
```

```python
df.shape
```

```python
df=df.fillna(0)
df
```

```python
df=df.astype("int64")
```

```python
df=df.fillna(0)
```

```python
df["Q1"] = df["Q1aM4"] + df["Q1bM6"]
df["Q2"] = df["Q2aM6"] + df["Q2bM4"]
df["Q3"] = df["Q3aM5"] + df["Q3bM5"]
df["Q4"] = df["Q4aM3"] + df["Q4bM7"]
df["Q5"] = df["Q5M10"]
df["Q6"] = df["Q6aM4"] + df["Q6bM6"]
```

```python
df
```

```python
df.drop(["Q1aM4","Q1bM6","Q2aM6","Q2bM4","Q3aM5","Q3bM5","Q4aM3","Q4bM7","Q5M10","Q6aM4","Q6bM6"],axis=1, inplace=True)
df
```

```python
df=df.fillna(0)
```

```python
df.plot()
plt.show()
```

```python
plt.hist(df['Q1'],color='purple',edgecolor='black',bins=9)
plt.xlabel("Marks scored in Q1")
plt.ylabel("No.of students")
plt.title("Histogram of Q1")
```

```python
df.plot.scatter(x = 'Q1', y ='Total',color='red',s=40)
plt.title("Scatter plot for Q1")
```

```python
df.boxplot(by='Q4',column =['Total'], grid = False,color='black')
plt.title("Boxplot of Q4")
plt.ylabel("Total")
plt.show()
```

```python
n = pd.DataFrame({'1.1': [10, 20, 30]}, index=['A', 'B', 'C'])
n['1.1'].plot(kind='pie',figsize=(8,8))
plt.title("piechart of marks in Q5")
plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Increase figure size
plt.plot(df["Q2"], marker='o', linestyle='-', color='b')
plt.title("Trend of Q2 Marks", fontsize=14)
plt.xlabel("Student Index", fontsize=12)
plt.ylabel("Total Marks", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True)  
plt.show()

```

```python
plt.figure(figsize=(12, 6))
df.iloc[:, 1:].head(10).plot(kind="bar", stacked=True, colormap="viridis", ax=plt.gca())
plt.title("Stacked Bar Chart (First 10 Students)", fontsize=14)
plt.xlabel("Student Index", fontsize=12)
plt.ylabel("Marks", fontsize=12)
plt.xticks(rotation=360)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap of Score Correlations", fontsize=14)
plt.xticks(rotation=360, ha="right", fontsize=10)
plt.yticks(rotation=0, fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.fill_between(range(len(df)), df["Total"].cumsum(), color='cyan', alpha=0.5)
plt.plot(df["Total"].cumsum(), marker='o', linestyle='-', color='blue', alpha=0.7)
plt.title("Cumulative Sum of Total Marks", fontsize=14)
plt.xlabel("Student Index", fontsize=12)
plt.ylabel("Cumulative Marks", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(8, 5))

questions = ["Q1", "Q2", "Q3", "Q4", "Q5", "Q6"]
means = df[questions].mean()
std_devs = df[questions].std()

plt.errorbar(questions, means, yerr=std_devs, fmt='o', color='red', capsize=5, markersize=8, linewidth=2)
plt.title("Mean & Standard Deviation of Each Question", fontsize=14)
plt.xlabel("Questions", fontsize=12)
plt.ylabel("Mean Score", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```
